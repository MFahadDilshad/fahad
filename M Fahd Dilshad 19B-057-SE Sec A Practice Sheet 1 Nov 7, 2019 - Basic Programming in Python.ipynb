Q5: A football is kicked 1 meter from the ground at an angle of 30 degree with a velocity of 15 m/sec.

<br>a. Max height above the ground.
<br>b. Total time in the air.
<br>c. Total distance traveled.
<br>d. Velocity vector at max height.
<br>e. Velocity when the ball hits the ground.

![proj](proj.jpg)



```python
from math import sin
#Data
v=15
g=-9.8
ang=30
distance=1

# Max height
h = (v * sin(ang)/g)+1
print("Max height is:",h,"meter")

print(" ")
#Total time in air
t = 2*v*sin(2*ang)/g
print("total time in air is:",t,"s")
print(" ")
#Total distance traveled.
d = v**2 * (sin(2*ang))/g
print("total distance travelled is:",d,"m")

print(" ")

#Velocity vector at max height.
Vx = v* distance
print("velocity vector at max height is:",Vx)

print(" ")

#Velocity when the ball hits the ground
Vf= h*(-g)
print("Velovity when the ball hits the ground is:",Vf,"m/s")
```
